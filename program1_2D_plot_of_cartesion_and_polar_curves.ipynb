Circle:x^2+y^2=a^2 for a=2

```python
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p2=plot_implicit(Eq(x**2+y**2,4),(x,-5,5),(y,-5,5),line_color='green',title='circle:x^2+y^2=4')
```

2) Strophoid: y^2(a-x)=x^2(a+x),a=3

```python
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p2=plot_implicit(Eq(y**2*(3-x),x**2*(3+x)),(x,-5,5),(y,-5,5),line_color="green",title='strophoid:y^2(a-x)=x^2(a+x),a=3')
```
