Circle:x^2+y^2=a^2 for a=2

```python
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p2=plot_implicit(Eq(x**2+y**2,4),(x,-5,5),(y,-5,5),line_color='green',title='circle:x^2+y^2=4')
```

2) Strophoid: y^2(a-x)=x^2(a+x),a=3

```python
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p2=plot_implicit(Eq(y**2*(3-x),x**2*(3+x)),(x,-5,5),(y,-5,5),line_color="green",title='strophoid:y^2(a-x)=x^2(a+x),a=3')
```

3) Cissoid:y^2(a-x)=x^3,a=2

```python
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p2=plot_implicit(Eq(y**2*(2-x),x**3),(x,-5,5),(y,-5,5),line_color="green",title='cissoid:y^2(a-x)=x^3')
```

4)Lemniscate:a^2*y^2=x^2(a^2-x^2),a=2

```python
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p2=plot_implicit(Eq(2**2*y**2,x**2*(2**2-x**2)),(x,-5,5),(y,-5,5),line_color='green',title='lemniscate:$a^2*y^2=x^2(a^2-x^2)$')
```

5)Astroid:x^(2/3)+y^(2/3)=a^(2/3),a=2

```python
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p2=plot_implicit(Eq(x**(2/3)+y**(2/3),2**(2/3)),(x,-5,5),(y,-5,5),line_color='green',title='astroid:$x^(2/3)+y^(2/3)=a^(2/3)$')
```

6)Folium of De-cartes:x^3+y^3=3axy,a=2

```python
from sympy import plot_implicit,symbols,Eq
x,y=symbols('x y')
p2=plot_implicit(Eq(x**3+y**3,3*2*x*y),(x,-5,5),(y,-5,5),line_color='green',title='Folium of De-cartes=x^3+y^3=3axy')
```

Plot Polar curves

7) Circle:x=acos(theta),y=bsin(theta),a=2,b=2

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,100)
x=2*cos(theta)
y=2*sin(theta)
r=sqrt(x**2+y**2)
polar(theta,r,'r+')
show()
```

8) Cardioids r=a(1+cos(theta)),r=a(1-cos(theta)),a=2

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,100)
r1=2*(1+cos(theta))
polar(theta,r1,'b+')
r2=2*(1-cos(theta))
polar(theta,r2,'r--')
show()
```

9)Four leaved rose r=2(cos(2theta))

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,100)
r1=2*abs(cos(2*theta))
polar(theta,r1,'b--')
show()
```

10) spiral of archimedes r=a+b(theta),a=2,b=3

```python
import numpy as np
from pylab import *
theta=linspace(0,2*np.pi,100)
r1=2+3*theta
polar(theta,r1,'b--')
show()
```
